## Importa Bibliotecas

```python
import google.generativeai as genai #Faz conex√£o com as llm do google
from google.colab import userdata #Faz conex√£o com ambiente colab do usu√°rio

#Pacotes de Formata√ß√£o
import textwrap
from IPython.display import Markdown
```

## Configura√ß√£o de Chave
1. V√° em https://aistudio.google.com/api-keys, crie a chave e copie.
2. No colab, crie uma vari√°vel de ambiente clicando no s√≠mbolo de "üîë" na esquerda do seu monitor. Cole a chave em 'value' e o nome deve ser igual ao que vc colocar no lugar de GEMINI_API_KEY entre aspas no c√≥digo abaixo.

```python
GOOGLE_API_KEY = userdata.get("GEMINI_API_KEY") #Do usu√°rio, pega "get" a vari√°vel com determinado nome.

genai.configure(api_key=GOOGLE_API_KEY) #Para acessar o gemini, configura a chave de api aquela guardada na vari√°vel.
```

## Escolher Modelo
Digite o nome do modelo entre as aspas.

```python
model = genai.GenerativeModel("gemini-2.5-flash")
```

```python
#Execute esse bloco APENAS se quiser ver uma lista de modelos.
for m in genai.list_models():
    if "generateContent" in m.supported_generation_methods:
        print(m.name)
```

## Enviar Pergunta
Coloque a pergunta que voc√™ quer fazer entre as aspas.

```python
%%time
response = model.generate_content(
    "Explique passo a passo, com 3 exemplos, a diferen√ßa entre classe e objeto em pyhton, com a s√≠ntaxe de cada um e de maniera did√°tica."
    )
```

## Configura√ß√µes para Formata√ß√£o
A fun√ß√£o abaixo vai formatar melhor a respota, para deixar mais "limpa" visualmente.

```python
def to_markdown(text):
    text = text.replace("‚Ä¢", "  *")
    return Markdown(textwrap.indent(text, "> ", predicate=lambda _: True))
```

Exibir Resposta

```python
to_markdown(response.text)
#Teste print(response.txt) para ver a diferen√ßa sem a formata√ß√£o.
```
